```python
import pandas as pd
import numpy as np
import os
```

```python
 #Read all csv in Fiscal_Data_Raw folder 
folder_path = '../Data/FisCal_Data_Raw'
all_files = os.listdir(folder_path)
csv_files = [f for f in all_files if f.endswith('.csv')]
```

```python
#combine csv files

combined_df = pd.concat(
    [pd.read_csv(os.path.join(folder_path, f)) for f in csv_files], ignore_index=True)
```

```python
combined_df.head()
```

```python
# Final_df (spending only from 2023 budget act)
final_df = combined_df[
    (combined_df['budget_reference_sub_category'] == "Budget Act") &
    (combined_df['year_of_enactment'] == 2023)
]
```

```python
#final_df to csv
final_df.to_csv('../Data/Fiscal_Data_Final.csv', index=False)
```
